```python
import json
def extract_genre_names(metadata):
    # 각 리스트에서 'name' 키 값을 추출
    corrected_json_string = metadata.replace("'", '"')
    json_data = json.loads(corrected_json_string)
    return [genre['name'] for genre in json_data if 'name' in genre]
```

```python

movie_metadata['genre_names']= movie_metadata['genres'].apply(extract_genre_names)

all_genres = set([genre for genres in movie_metadata['genre_names'] for genre in genres])

# 더미 변수 컬럼 만들기
for genre in all_genres:
    movie_metadata[genre] = movie_metadata['genre_names'].apply(lambda x: 1 if genre in x else 0)
```

```python
movie_metadata.columns
X = movie_metadata[['adult', 'budget', 'id', 'popularity', 'revenue', 'runtime',
       'status', 'video', 'vote_average', 'vote_count', 'TV Movie', 'Animation', 'Adventure', 'Music', 'Foreign', 'Action',
       'War', 'Comedy', 'Fantasy', 'Science Fiction', 'Crime', 'Romance',
       'History', 'Thriller', 'Horror', 'Family', 'Mystery', 'Western',
       'Drama', 'Documentary']]

X = pd.get_dummies(X, columns=['status'])
X
```

```python
#X[X['popularity'].isnull()]
X = X.dropna(subset='popularity').reset_index()
X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
# plt.savefig('images/11_03.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=3,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
y_km = kmeans.fit_predict(X)

silhouette_avg = silhouette_score(X, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg}')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=2,
                init='k-means++',
                n_init='auto',
                max_iter=100,
                random_state=0)
y_km = kmeans.fit_predict(X)
# 클러스터링된 결과 (K-Means 등으로 미리 클러스터링했다고 가정)
clusters = y_km  # 예시 클러스터 레이블
centroids = kmeans.cluster_centers_
# PCA로 2차원으로 축소 (고차원 데이터를 2D로 변환)
pca = PCA(n_components=2)
reduced_movies = pca.fit_transform(X)
reduced_centroids = pca.transform(centroids)


# 색상 팔레트 설정 (클러스터별로 색상을 구분)
palette = sns.color_palette("hsv", np.unique(clusters).shape[0])

# 그래프 그리기
plt.figure(figsize=(8, 6))
sns.scatterplot(x=reduced_movies[:, 0], y=reduced_movies[:, 1], hue=clusters, palette=palette, s=100)
plt.scatter(reduced_centroids[:, 0], reduced_centroids[:, 1], s=200, c='black', marker='X', label='Centroid')

# 그래프 제목 및 축 레이블
plt.title('Clustered Movies')
plt.xlabel('PCA Component 1 (Genre Similarity)')
plt.ylabel('PCA Component 2 (Popularity)')
plt.legend(title='Cluster')
plt.grid(True)

# 그래프 출력
plt.show()
```

```python
# 클러스터별 대중성을 기준으로 그룹화했을때 값의 비교
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import euclidean_distances

def get_cluster_popularity(n_cluster):
    scaler = MinMaxScaler()

    cluster_indices = np.where(y_km == n_cluster)[0]
    cluster_data = X.loc[cluster_indices]

    popularity_data =  cluster_data[['popularity','revenue']]
    # popularity_data
    scaled_popularity = scaler.fit_transform(popularity_data)

    # # # 유클리디안 거리 계산 (거리가 작을수록 대중성이 유사함)
    similarity_matrix = euclidean_distances(scaled_popularity)
    movie_ids = cluster_data['id']
    similar_movie_ids = get_movie_id(similarity_matrix,movie_ids, 0.8)
    # similar_movie_ids

    all = movie_metadata[['id','title','original_title','genres']].merge(cluster_data, how='inner', on='id')
    # all['genres']= all['genres'].apply(extract_genre_names)
    return all[all['id'].isin(similar_movie_ids)][['popularity','revenue']]
    
```
