```python
from ast import literal_eval

features = ['cast','crew','genres','keywords','spoken_languages','belongs_to_collection']

for feature in features:
    movies_metadata[feature] = movies_metadata[feature].apply(
    lambda x: literal_eval(x) if pd.notna(x) else 'null'
)
```

```python
def extract_name(rows, attribute = 'name'):
    if rows is None:
        return np.nan
    if isinstance(rows, list):
        if len(rows) == 0:
            return np.nan
        #return [row[attribute] if attribute in row else np.nan for row in rows]
        return ",".join(row.get(attribute, ' ') for row in rows)
    return np.nan
```

```python
movies_metadata['genres'] = movies_metadata['genres'].apply(extract_name)
```

```python
movies_metadata['spoken_languages'] = movies_metadata['spoken_languages'].apply(lambda x: extract_name(x,'iso_639_1'))
```

```python
movies_metadata['keywords'] = movies_metadata['keywords'].apply(extract_name)
```

```python
movies_metadata['cast'] = movies_metadata['cast'].apply(extract_name)
```

```python
def get_director(x):
    if x is None:
        return np.nan
    if isinstance(x, list):
        for i in x:
            if isinstance(i, dict) and 'job' in i and i['job'] == 'Director':
                return i['name']
    return np.nan
movies_metadata['director'] = movies_metadata['crew'].apply(get_director)
```

```python
movies_metadata['belongs_to_collection'] = movies_metadata['belongs_to_collection'].apply(lambda x: x['name'] if pd.notna(x) and 'name' in x else x)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies = movies_metadata

# 'overview', 'genres', 'keywords' 컬럼을 합쳐서 하나의 텍스트로 만든다.
movies['combined'] = movies['overview'] + " " + movies['genres'] + " " + movies['keywords'] + movies['belongs_to_collection'] + movies['director']
movies['combined'] = movies['combined'].fillna('')
# TF-IDF 벡터화
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['combined'])

# # # 코사인 유사도 계산
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

```

```python
def recommend_movie(movie_title, cosine_sim=cosine_sim):
    # 영화 제목의 인덱스를 찾는다
    #print(movies.head())
    #print(movies[movies['original_title'] == movie_title])
    idx = movies.index[movies['original_title'] == movie_title].tolist()[0]
    
    # 해당 영화와 다른 영화들 간의 유사도를 가져온다
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 유사도를 기준으로 내림차순 정렬하고, 첫 번째 영화(자기 자신 제외) 추천
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:]
    
    # 가장 유사한 3개의 영화 가져오기
    movie_indices = [i[0] for i in sim_scores[:3]]
    
    # 추천 영화 리스트
    recommended_movies = movies['original_title'].iloc[movie_indices]
    
    return recommended_movies

# 'Toy Story'를 기준으로 추천
recommended = recommend_movie('Toy Story')
print(recommended)
```
